```python
import json
import pandas as pd
import matplotlib.pyplot as plt

with open("/workspace/FIA1471/shopee_log/shopee_data_20241028_023517.json") as f:
    data = json.load(f)

# Extract products data
products = data['data']['sections'][0]['data']['item']

# List essential fields for EDA
fields = [
    'itemid', 'shopid', 'name', 'price', 'price_before_discount', 'stock', 'sold',
    'historical_sold', 'liked_count', 'cmt_count', 'rating_star', 'rating_count',
    'shop_location', 'shop_name', 'catid', 'brand'
]

# Extracting and structuring data
product_data = [{field: product.get(field, None) for field in fields} for product in products]
df = pd.DataFrame(product_data)

# Quick EDA
print(df.info())
print(df.describe())

# Example visualizations for EDA
# plt.figure(figsize=(10, 6))
# plt.hist(df['price'] / 1e6, bins=30)  # price in millions for readability
# plt.title('Price Distribution (in Million VND)')
# plt.xlabel('Price (Million VND)')
# plt.ylabel('Frequency')
# plt.show()

# plt.figure(figsize=(10, 6))
# plt.scatter(df['sold'], df['price'] / 1e6, alpha=0.5)
# plt.title('Sales vs Price')
# plt.xlabel('Total Sales')
# plt.ylabel('Price (Million VND)')
# plt.show()
```
